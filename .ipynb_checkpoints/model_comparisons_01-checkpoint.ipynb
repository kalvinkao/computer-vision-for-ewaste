```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
classes = ['tv','laptop','mouse','remote','keyboard','cell']
models = ['ssd','faster rcnn', 'yolo']
overallMAP = [30.5, 31.6, 37.1]
electronicsMAP = [36.1, 36.0, 41.7]
avg_pct_unique_electronics_labels_IDed_per_pic = [57.8, 74.2, 57.1]
```

```python
barWidth = 0.35
r1 = np.arange(len(models))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
```

```python
plt.bar(r1, overallMAP, width=barWidth, edgecolor='white', label='Overall')
plt.bar(r2, electronicsMAP, width=barWidth, edgecolor='white', label='Electronics Only')
plt.xlabel('Model', fontweight='bold')
plt.xticks([r + barWidth/2 for r in range(len(models))],models)
plt.title('MAP Score', fontweight='bold')
plt.legend()
plt.show()
```

```python
TPs_averaged_across_classes = [61.5, 77.0, 61.8]
overall_TPs = [61.3, 76.4, 60.5]
```

```python
barWidth = 0.4
plt.bar(r1, avg_pct_unique_electronics_labels_IDed_per_pic, width=barWidth, edgecolor='white',label='by Picture')
#plt.bar(r2, TPs_averaged_across_classes, width=barWidth, edgecolor='white')
plt.bar(r2, overall_TPs, width=barWidth, edgecolor='white', label='Overall')
plt.xlabel('Model', fontweight='bold')
plt.xticks([r + barWidth/2 for r in range(len(models))],models)
plt.title('Percentage of Unique Labels Identified')
#plt.legend(loc='lower center')
plt.legend()
plt.show()
```

```python
yolo_TPs = [0.691, 0.705, 0.747, 0.493, 0.641, 0.429]
rcnn_TPs = [0.884, 0.8196, 0.8295, 0.586, 0.849, 0.6495]
ssd_TPs = [0.787, 0.758, 0.670, 0.3958, 0.6604, 0.4171]
```

```python
barWidth = 0.2
s1 = np.arange(len(classes))
s2 = [x + barWidth for x in s1]
s3 = [x + barWidth for x in s2]
```

```python
plt.figure(figsize=(9,6))
plt.bar(s1, ssd_TPs, width=barWidth, edgecolor='white', label='ssd')
plt.bar(s2, rcnn_TPs, width=barWidth, edgecolor='white', label='faster rcnn')
plt.bar(s3, yolo_TPs, width=barWidth, edgecolor='white', label='yolo')
plt.xlabel('Classes', fontweight='bold')
plt.xticks(s2,classes)
plt.title('Fraction of Images in which Label was Detected', fontweight='bold')
plt.legend()
plt.show()
```
